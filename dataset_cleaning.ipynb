```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
original_data = pd.read_csv("MushroomDataset/secondary_data.csv", sep=";")
```

```python
original_data.head()
```

```python
original_data.isna().sum()
```

```python
original_data["class"].unique()
```

```python
data_occurences = original_data.groupby("class").size().reset_index().sort_values(by=0, ascending=False)
data_occurences = data_occurences.set_axis(["class", "value"], axis=1)
data_occurences["%total"] = [x*100/original_data["class"].shape[0] for x in data_occurences["value"]]
data_occurences
```
